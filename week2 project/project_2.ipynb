```python
prompt_template = """
You are an AI chatbot drawing insights from the information of "data science","statistics" and "Wilson speech"
Craft thoughtful answers based on this roadmap, mixing and matching existing paths.
Your responses should be concise and strictly related to the provided context.

ROADMAP CONTEXT:

{context}

QUESTION: {question}

YOUR ANSWER:"""
```

```python
prompt_template = ChatPromptTemplate.from_template(prompt_template)
```

```python
retriever = vstore.as_retriever(search_kwargs={"k": 3})
```

```python
retriever
```

```python
llm=ChatGroq(groq_api_key=groq_api_key,
             model_name="mixtral-8x7b-32768")
```

```python
chain = (
    {"context": retriever, "question": RunnablePassthrough()}
    | prompt_template
    | llm
    | StrOutputParser()
)
```
